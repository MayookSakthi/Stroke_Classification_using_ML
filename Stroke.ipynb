```python
target='stroke'
x=df.loc[:,df.columns!=target]
y=df.loc[:,df.columns==target]
```

our target value is stroke
so we give target value into y and we give other features into X
in our dataset 11 features. one is dependent variable and 10 features are independent variable  

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics 
```

# Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=200,test_size=0.20)  
```

we assign 2:8 ratio size test and train data

```python
dtree=DecisionTreeClassifier(max_depth=300)
```

```python
dtree_2=dtree.fit(x_train,y_train)
```

```python
y_pred=dtree_2.predict(x_test)
```

```python
print("Accuracy :",metrics.accuracy_score(y_test,y_pred)*100)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=123,
                                               test_size=0.20)
```

```python
rf=RandomForestClassifier(n_estimators=600)
```

```python
rf=rf.fit(x_train,y_train)
```

```python
y_predic=rf.predict(x_test)
```

```python
print('Accuracy:',metrics.accuracy_score(y_test,y_predic)*100)
```
